**Extract the currencies data from the Countries json file**

```python
import pandas as pd

df_currencies = pd.DataFrame(columns = ['Currencies'])

for country_id, country_info in data.items():
    
    #Remove extra white spaces
    country_info['currency'] = str(country_info['currency']).replace(' ','')
    #Replace apostrophe with empty space
    country_info['currency'] = str(country_info['currency']).replace("'", " ")
    #Replace brackets with empty space
    country_info['currency'] = str(country_info['currency']).replace("[", " ")
    country_info['currency'] = str(country_info['currency']).replace("]", " ")
    
    df_currencies.loc[len(df_currencies)] = country_info['currency']
    #Remove extra white spaces in the dataframe
    df_currencies['Currencies'] = df_currencies['Currencies'].replace(' ','')

df_currencies
```

**Create table and load continents data into postgresql database**

```python
import psycopg2
import pandas as pd
import numpy as np
import psycopg2.extras as extras

#A function that inserts the data in the pandas dataframe into a postgresql database
def execute_values(conn, df, table, Query):
  
    tuples = [tuple(x) for x in df.to_numpy()]
  
    cols = ','.join(list(df.columns))
    query2 = "INSERT INTO %s(%s) VALUES %%s" % (table, cols)
    cursor = conn.cursor()
    try:
        cursor.execute(Query)
        extras.execute_values(cursor, query2, tuples)
        conn.commit()
    except (Exception, psycopg2.DatabaseError) as error:
        print("Error: %s" % error)
        conn.rollback()
        cursor.close()
        return 1
    print("the dataframe is inserted")
    cursor.close()
  
  
#Establishing the connection to the postgressql database named test
conn = psycopg2.connect(
    database = "assignment", user = "postgres", host = 'localhost', port = "5432"
)
  
df = pd.DataFrame(Continents)
#renaming the continents code with actual name of the continents
rename = {'AF':'Africa', 'AN':'Antartica','AS':'Asia','EU':'Europe','NA':'North America','OC':'Oceania','SA':'South America'}
df = df.replace(rename)
df.columns = ["Continents"]
df = df.replace(np.nan,None,regex=True)

#Executing the Function
execute_values(conn, df, 'continents','CREATE TABLE continents (continents VARCHAR(50))')
```

**Add unique ids to each table and make them primary keys**

```python
import psycopg2
import pandas as pd
import numpy as np
import psycopg2.extras as extras

conn = psycopg2.connect(
    database = "assignment", user = "postgres", host = 'localhost', port = "5432"
)

cursor = conn.cursor()
try:
    cursor.execute('''ALTER TABLE continents ADD COLUMN id SERIAL PRIMARY KEY;
    
    ''')

    cursor.execute('''ALTER TABLE countries ADD COLUMN id SERIAL PRIMARY KEY;
    
    ''')

    cursor.execute('''ALTER TABLE currencies ADD COLUMN id SERIAL PRIMARY KEY;
    
    ''')
    cursor.execute('''ALTER TABLE languages ADD COLUMN id SERIAL PRIMARY KEY;
    
    ''')
    conn.commit()
except (Exception, psycopg2.DatabaseError) as error:
    print("Error: %s" % error)
    conn.rollback()
    cursor.close()
# print("the dataframe is inserted")
cursor.close()
```
